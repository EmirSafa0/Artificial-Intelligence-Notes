```python
def OrtalamaIleDoldur(data):
  toplam=data.sum().sum()  # Data daki verilerin hepsini toplar.
  elemanSayisi=data.size-data.isnull().sum().sum()  # Data boyutu (6*6) - data daki NaN degerler (6)
  ortalama=toplam/elemanSayisi
  return ortalama
```

```python
data.fillna(OrtalamaIleDoldur(data))
```

```python
data.mean(skipna=True)  # O sutundaki NaN olmayan butun verileri toplar ve NaN olmayan eleman sayisina boler.
# skipna=True NaN degerleri atlamamizi saglar.
```

```python
from sklearn.impute import SimpleImputer  # NaN degerleri doldurmak icin kullanilan kutuphanedir.
eksikVeriTamamlama=SimpleImputer(missing_values=np.nan,strategy="mean") # missing_values=np.nan NaN verileri doldurcak. strategy="mean" eksik verileri sutun ortalamasiyla doldurur. Bu satirda eksikler doldurulmuyor.
eksikVeriTamamlama=eksikVeriTamamlama.fit(data.iloc[:,1:4]) # fit ile NaN olmayan degerlerin ortalamasi hesaplanir hafizada tutulur.
data.iloc[:,1:4] = eksikVeriTamamlama.transform(data.iloc[:,1:4]) # transform fit'i kullanarak NaN'lari doldurur.
data
```

```python
data = pd.read_csv("../data/raw/eksikveriler2.csv")

from sklearn.impute import SimpleImputer
eksikVeriTamamlama=SimpleImputer(missing_values=np.nan,strategy="median") # Medyanla doldurur.
eksikVeriTamamlama=eksikVeriTamamlama.fit(data.iloc[:,1:4])
data.iloc[:,1:4] = eksikVeriTamamlama.transform(data.iloc[:,1:4])
data
```

```python
data = pd.read_csv("../data/raw/eksikveriler2.csv")

from sklearn.impute import SimpleImputer
eksikVeriTamamlama=SimpleImputer(missing_values=np.nan,strategy="most_frequent") # En sik gorulen deger ile doldurur.
eksikVeriTamamlama=eksikVeriTamamlama.fit(data.iloc[:,1:4])
data.iloc[:,1:4] = eksikVeriTamamlama.transform(data.iloc[:,1:4])
data
```
