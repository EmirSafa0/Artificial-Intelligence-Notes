```python
import pandas as pd  # Veri okuma ve DataFrame islemleri icin pandas
import numpy as np  # Sayisal islemler icin numpy
import matplotlib.pyplot as plt  # Grafik cizimi icin matplotlib
import seaborn as sns  # Gelismis grafikler icin seaborn
from sklearn.model_selection import train_test_split  # Veriyi egitim ve test setlerine ayirmak icin
from sklearn.preprocessing import LabelEncoder  # Kategorik etiketleri sayisal hale getirmek icin
from sklearn.preprocessing import StandardScaler  # Ozellikleri standardize etmek icin
from sklearn.metrics import classification_report, confusion_matrix  # Rapor ve confusion matrix icin
from sklearn.linear_model import LogisticRegression  # Logistic Regression modeli
from sklearn.tree import DecisionTreeClassifier  # Decision Tree modeli
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier, GradientBoostingClassifier  # Ensemble modeller
from sklearn.neighbors import KNeighborsClassifier  # KNN modeli
from sklearn.svm import SVC  # SVM modeli
from sklearn.naive_bayes import GaussianNB  # Naive Bayes modeli

# veri seti yukleme

data_yolu = "../data/raw/data.csv"  # Veri setinin dosya yolunu belirtir
data = pd.read_csv(data_yolu)  # CSV dosyasini okuyarak DataFrame haline getirir

# onislem

data.drop(["id", "Unnamed: 32"], axis=1, inplace=True)  # Gereksiz sutunlari kalici olarak siler
label_encoder = LabelEncoder()  # LabelEncoder nesnesini olusturur
data["diagnosis"] = label_encoder.fit_transform(data["diagnosis"])  # diagnosis etiketlerini 0/1 seklinde sayisallastirir
X = data.drop("diagnosis", axis=1)  # diagnosis sutununu cikartarak ozellikleri (features) olusturur
y = data["diagnosis"]  # diagnosis sutununu hedef degisken (etiket) olarak alir
scaler = StandardScaler()  # StandardScaler nesnesini olusturur
X_scaled = scaler.fit_transform(X)  # Tum ozellikleri standardize eder (ortalama 0, std 1)
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.3, random_state=42, stratify=y
)  # Veriyi %70 egitim %30 test olacak sekilde ve sinif dagilimini koruyarak ayirir

# model kurulumu

models = {  # Denenecek modelleri isim:model seklinde sozlukte toplar
    "Logistic Regression": LogisticRegression(),  # Logistic Regression modelini ekler
    "Decision Tree": DecisionTreeClassifier(),  # Decision Tree modelini ekler
    "Random Forest": RandomForestClassifier(),  # Random Forest modelini ekler
    "Gradient Boost": GradientBoostingClassifier(),  # Gradient Boosting modelini ekler
    "AdaBoost": AdaBoostClassifier(),  # AdaBoost modelini ekler
    "KNN": KNeighborsClassifier(),  # KNN modelini ekler (varsayilan k=5)
    "SVM": SVC(),  # SVM modelini ekler
    "Naive Bayes": GaussianNB()  # Gaussian Naive Bayes modelini ekler
}

# model egitimi ve sonuclar

dogruluk_listesi = []  # Her modelin accuracy degerini saklamak icin bos liste
for model_name, model in models.items():  # Sozlukteki her modeli tek tek dolasir
    model.fit(X_train, y_train)  # Modeli egitim verisi ile egitir
    y_pred = model.predict(X_test)  # Test verisi icin tahmin yapar
    dogruluk = model.score(X_test, y_test)  # Test seti uzerinden accuracy hesaplar
    dogruluk_listesi.append((model_name, dogruluk))  # Model adi ve accuracy degerini listeye ekler
    cm = confusion_matrix(y_test, y_pred)  # Confusion matrix olusturur
    print(f"{model_name}-Siniflandirma Raporu:\n{classification_report(y_test, y_pred, target_names=['Benign', 'Malignant'])}")   # Her model icin precision/recall/f1-score raporunu yazdirir
    plt.figure(figsize=(6, 4))  # Confusion matrix grafiginin boyutunu ayarlar
    sns.heatmap(
        cm, annot=True, fmt="d", cmap="Blues",
        xticklabels=["Benign", "Malignant"], yticklabels=["Benign", "Malignant"]
    )  # Confusion matrixi heatmap olarak cizer
    plt.xlabel("Predicted")  # X ekseni adini belirler
    plt.ylabel("True")  # Y ekseni adini belirler
    plt.title(f"{model_name}-Confusion Matrix")  # Grafik basligini belirler
    plt.show()  # Grafigi ekranda gosterir

# dogruluk degerleri cizdir

dogruluk_df = pd.DataFrame(dogruluk_listesi, columns=["Model", "Accuracy"])  # Model-accuracy listesinden DataFrame olusturur
plt.figure(figsize=(10, 6))  # Bar grafigin boyutunu ayarlar
sns.barplot(x="Model", y="Accuracy", data=dogruluk_df, palette="viridis")  # Modellerin accuracy degerlerini bar grafikle cizer   *** viridis ise mordan sariya gecen, okunmasi kolay bir renk paletidir.
plt.xlabel("Model")  # X ekseni adini belirler
plt.ylabel("Accuracy")  # Y ekseni adini belirler
plt.title("Model Dogruluklari")  # Grafigin basligini belirler
plt.xticks(rotation=90)  # Model isimlerini dikey okumak icin 90 derece dondurur
plt.show()  # Grafigi ekranda gosterir
```
