```python
iris.iloc[0, 1] = np.nan  # Iris DataFrame'inde 0. satir 1. sutundaki degere NaN (eksik veri) atar
iris.iloc[0, 2] = np.nan  # Iris DataFrame'inde 0. satir 2. sutundaki degere NaN (eksik veri) atar

iris.isnull().sum()  # Her sutunda kac tane NaN (eksik veri) oldugunu hesaplar ve gosterir

iris_ortalama = iris.copy()  # Orijinal iris verisinin ortalama ile doldurma icin kopyasini olusturur
iris_medyan = iris.copy()    # Orijinal iris verisinin medyan ile doldurma icin kopyasini olusturur
iris_mod = iris.copy()       # Orijinal iris verisinin mod ile doldurma icin kopyasini olusturur

iris_ortalama.fillna(iris.mean(numeric_only=True), inplace=True)  # Sayisal sutunlardaki NaN degerleri sutun ortalamalari ile doldurur
iris_medyan.fillna(iris.median(numeric_only=True), inplace=True)  # Sayisal sutunlardaki NaN degerleri sutun medyanlari ile doldurur
iris_mod.fillna(iris.mode(numeric_only=True).iloc[0], inplace=True)  # Sayisal sutunlardaki NaN degerleri sutun modlari ile doldurur
```

```python
from sklearn.ensemble import RandomForestClassifier  # Random Forest siniflandirma algoritmasini ice aktarir
from sklearn.model_selection import train_test_split # Veriyi egitim ve test olarak ayirmak icin kullanilir
from sklearn.metrics import accuracy_score            # Modelin dogruluk oranini hesaplamak icin kullanilir
from sklearn.metrics import classification_report     # Precision, recall, f1-score gibi metrikleri hesaplar

dosya_yolu = "../data/raw/iris.csv"  # Iris veri setinin bulundugu dosya yolunu belirtir
iris = pd.read_csv(dosya_yolu)    # CSV dosyasini okuyarak DataFrame haline getirir

x = iris.drop(["Id", "Species"], axis=1)  # Bagimsiz degiskenleri olusturur (ozellikler)
y = iris["Species"]                       # Bagimli degiskeni olusturur (sinif etiketi)

x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.30, random_state=0
)
# Veriyi %70 egitim, %30 test olacak sekilde ayirir
# random_state=0 ayni bolmenin tekrar edilmesini saglar

model = RandomForestClassifier()  # Random Forest siniflandirici modelini olusturur
model.fit(x_train, y_train)       # Modeli egitim verileri ile egitir

y_pred = model.predict(x_test)    # Test verileri icin sinif tahminleri yapar

dogruluk = accuracy_score(y_test, y_pred)  # Gercek ve tahmin edilen siniflara gore dogruluk oranini hesaplar
print("Dogruluk Orani:", dogruluk)          # Dogruluk oranini ekrana yazdirir

rapor = classification_report(y_test, y_pred)  # Siniflandirma performans raporunu olusturur
print("Siniflandirma Raporu:\n", rapor)         # Precision, recall ve f1-score degerlerini ekrana yazdirir
```
